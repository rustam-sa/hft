```python
ds = DataService()
data = ds.get_dataframes_as_dicts(collection_name='image_design_save_test')
```

```python

import pandas as pd
import image_maker
# Check timestamps
timestamps = pd.Series([s['timestamp'] for s in data])
checks = image_maker.check_timestamps(timestamps)
if not checks['ascending'] or checks['has_gaps']:
    print(checks)
    raise ValueError("Timestamps are not in ascending order or there are gaps larger than 3 minutes.")

# Split data
train, val, test = image_maker.split_list(data, fractions=fractions)
check = image_maker.check_temporal_order(train, val, test)
if not check:
    print(checks)
    raise ValueError("Train, validation, and test sets are not properly temporally separated.")
```
